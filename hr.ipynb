```python
import matplotlib.pyplot as plt
import seaborn as sns
# Distribution of Age, Degree Percentage, Experience, Salary
plt.figure(figsize=(15, 6))
for i, col in enumerate(['Age', 'Degree Percentage', 'experience', 'Salary'], 1):
    plt.subplot(1, 4, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Average Salary and Experience per Department
dept_stats = df.groupby('Department')[['Salary', 'experience']].mean()
print(dept_stats)

# Status analysis by Department
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Department', hue='Status')
plt.title('Status by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Age', 'Degree Percentage', 'experience', 'Salary']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Calculate Tenure
df['Tenure'] = 2024 - df['Year of Joining']
sns.scatterplot(data=df, x='Tenure', y='Salary', hue='Status')
plt.title('Tenure vs Salary by Status')
plt.show()
```
